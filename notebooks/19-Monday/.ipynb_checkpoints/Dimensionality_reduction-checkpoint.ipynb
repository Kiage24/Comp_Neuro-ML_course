```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg
from sklearn.decomposition import PCA, FactorAnalysis
import scipy.io as spio
```


# Load data



```python
#Import data
datapath  = '/content/drive/MyDrive/Soldado-Magraner/TReND_CompNeuro/Teaching Materials/19-Monday/'
datafile  = 'data_for_exercises.mat'

mat       = spio.loadmat(datapath + datafile, squeeze_me=True)
Xplan     = mat['Xplan']  
data_dims = Xplan.shape
Xsim      = mat['Xsim'] 
sim_data_dims = Xsim.shape

print(data_dims)
print(sim_data_dims)
print('dims = (samples, features)')
```

# 1. Visualization of high-dimensional neural activity using PCA



# 2. From PCA to PPCA and FA

```python
# Run PPCA
pca = PCA(svd_solver="full")#,n_components ='mle')
pca.fit(Xsim)
print('Explained variance ratio')
print(pca.explained_variance_ratio_)
sigma2   = pca.noise_variance_
print('Noise variance')
print(sigma2)
ppca_noise_cov = sigma2 * np.eye(pca.n_features_in_)
PPCA_cov_eq = pca.components_.T * pca.components_ + ppca_noise_cov #pca.components dim=(num components,features)
#cov = components_.T * S**2 * components_ + sigma2 * eye(n_features) where S**2 contains the explained variances, and sigma2 contains the noise variances.
print('Data sample covariance')
print(np.cov(Xsim.T))
print('Data sample covariance approximated by PPCA, WW^T + sigma2*I')
print(PPCA_cov_eq)
print('Data sample covariance approximated by PPCA')
PPCA_cov = pca.get_covariance()
print(PPCA_cov)
z_hat_ppca = pca.transform(Xsim)  #dim=(samples,num components), returns PCA projections, but not PPCA ones            
```

```python
print(ppca_noise_cov)
```

```python
# Run FA
fa = FactorAnalysis(noise_variance_init=np.ones(2)*8)#rotation='varimax',
fa.fit(Xsim)
print('Explained variance ratio')
sigma2   = fa.noise_variance_
print('Noise variance')
print(sigma2)
fa_noise_cov = np.diag(sigma2)
FA_cov_eq = fa.components_.T * fa.components_ + fa_noise_cov #fa.components dim=(num components,features)
#cov = components_.T * S**2 * components_ + sigma2 * eye(n_features) where S**2 contains the explained variances, and sigma2 contains the noise variances.
print('Data sample covariance')
print(np.cov(Xsim.T))
print('Data sample covariance approximated by FA, WW^T + diag(sigma2)')
print(FA_cov_eq)
print('Data sample covariance approximated by FA')
FA_cov = fa.get_covariance()
print(FA_cov)
z_hat_fa = fa.transform(Xsim) #dim=(samples,num components)
```

Let's plot the results

```python
def dim_reduce_plot(x,z_hat,u1,title):
  # INPUTS:
  # x: high dimensional data, each row is a data point
  # z_hat: low dimensional data, each row is a data point
  # u1: principle direction / factor loading
  n = x.shape[0]
  p = x.shape[1]
  mu = np.mean(x,axis=0)
  #plot
  plt.plot(x[:,0],x[:,1],'o',color='black')
  plt.xlabel('dim 1')
  plt.ylabel('dim 2')
  plt.title(title)

  # plot the mean
  plt.plot(mu[0],mu[1],'o',color='green',markersize=15)
  scale = 2*np.max(np.std(x))/linalg.norm(u1);
  # plot the line defined by the first principal component
  xVals = scale*np.concatenate((-u1[0], u1[0])) # +/- 2 standard deviation
  yVals = scale*np.concatenate((-u1[1], u1[1])) # +/- 2 standard deviation
  plt.plot(mu[0]+xVals,mu[1]+yVals,linestyle='-',color='black')
  # plot the original data projected down into the subspace defined by PC1
  dataPC1 = np.tile(mu, (n, 1)) + np.tile(u1.T, (n, 1))*np.tile(z_hat,(1,p));
  plt.plot(dataPC1[:,0],dataPC1[:,1],'o',color='red')
  for i in range(n):
    xVals = np.array([x[i,0], dataPC1[i,0]])
    yVals = np.array([x[i,1], dataPC1[i,1]])
    plt.plot(xVals,yVals,linestyle='-',color='red');
    #axis equal
```

```python
# find the projection of the data onto PCA dim1
#z_hat_PCA = Xsim_centered*U[:,1]
#dim_reduce_plot(Xsim,z_hat_PCA,U[:,1])

# make the plot using the PPCA parameters from the sklearn package: this returns the PCA parameters...
#dim_reduce_plot(Xsim,z_hat_ppca,pca.components_.T,'PCA')
dim_reduce_plot(Xsim,z_hat_ppca[:,0,np.newaxis],pca.components_.T[:,0,np.newaxis],'PCA')
```

```python
# find the projection of the data onto PPCA dim1
MU   = np.mean(Xsim[:,:,np.newaxis],axis=0)
#C    = pca.components_.T * pca.components_ + ppca_noise_cov
Cinv = np.linalg.inv(PPCA_cov)
z_hat_ppca_eq = pca.components_ @ Cinv @ (Xsim.T-MU)

# make the plot using the PPCA parameters from the sklearn package: this returns the PCA parameters...
#dim_reduce_plot(Xsim,z_hat_ppca_eq.T,pca.components_.T,'PPCA')
dim_reduce_plot(Xsim,z_hat_ppca_eq.T[:,0,np.newaxis],pca.components_.T[:,0,np.newaxis],'PPCA')
```

```python
# find the projection of the data onto FA dim 1
MU   = np.mean(Xsim[:,:,np.newaxis],axis=0)
#C    = fa.components_.T * fa.components_ + fa_noise_cov
Cinv = np.linalg.inv(FA_cov)
z_hat_fa_eq = fa.components_ @ Cinv @ (Xsim.T-MU)

# plot
dim_reduce_plot(Xsim,z_hat_fa_eq.T[:,0,np.newaxis],fa.components_.T[:,0,np.newaxis],'FA')
#dim_reduce_plot(Xsim,z_hat_fa[:,0,np.newaxis],fa.components_.T[:,0,np.newaxis],'FA')
```
